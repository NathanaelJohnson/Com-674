```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set(style="ticks")
iris = sns.load_dataset("iris")
```

```python
# Exercise 1
# Visualize the distribution of another variable in the iris dataset

sns.distplot(iris.petal_length, kde=False, fit=stats.norm)
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# Exercise 2
# Plot a boxplot of the tip size by the size of the party.
sns.boxplot(x = 'size',
           y = 'tip',
           data=tips)
```

```python
# Exercise 3
# Take your code from Exercise 2, and color the boxplots by the gender of the tipper.
sns.boxplot(x = 'size',
           y = 'tip',
            hue = 'sex',
           data=tips)
```

```python
# Exercise 4
# Plot a swarm plot of the bill amount by the size of the party.
sns.swarmplot(x = 'size',
           y = 'total_bill',
           data=tips)
```

```python
# Exercise 5
# Now take your swarm plot and make a facet for each day of the week. Hint: To make it look nice, you need to figure out how to wrap it.
sns.catplot(x = 'size',
           y = 'total_bill',
                col = 'day', col_wrap = 2,
            kind='swarm',
           data=tips)
```

```python
# Exercise 6
# Come up with your own question about this dataset. Then, create a visualization that sheds light on that 
# question and make an argument about how it helps to answer the question.
# Do smokers or non-smokers tip better?

# This plot shows that, in this dataset, there appears to be an interaction effect, smokers tipped better 
# when the bill was less than qroun $18, but beyond that, non-smokers tipped better.
sns.lmplot(x = 'total_bill',
           y = 'tip',
                hue = 'smoker',
           data=tips)
```
